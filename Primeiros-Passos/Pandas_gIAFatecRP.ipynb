Seguindo abaixo temos informações sobre as colunas, tais como, quantidade (20), nome de cada coluna, seguido pela quantidade de valores presentes na coluna em questão, e seu tipo. Pro exemplo, a coluna _Net Migration_ possui 224 valores não nulos (_not null_) e é do tipo _object_. Sendo assim, é possível verificar que tal coluna possui 4 valores ausentes.

Entretanto, ao comparar a informação sobre as colunas com o arquivo em questão percebemos que grande parte das colunas deveriam ser do tipo _float_. Isto acontece devido ao formato dos números representados no arquivo com símbolo decimal __","__ e colocados entre aspas.

Para resolver tal problema o método _read_csv_ permite que seja passado por parâmetro funções de conversão para que as colunas tenham o formato adequado no Pandas. 


```
def removeSpaces(x):
  return x.strip()

def toFloat(x):
  return np.float(x.replace(',','.')) if x != '' else np.float('NaN')

convs = {'Country': removeSpaces,
         'Region': removeSpaces,
         'Pop. Density (per sq. mi.)': toFloat,
         'Coastline (coast/area ratio)': toFloat,
         'Net migration': toFloat,
         'Infant mortality (per 1000 births)': toFloat,
         'Literacy (%)': toFloat,
         'Phones (per 1000)': toFloat,
         'Arable (%)': toFloat,
         'Crops (%)': toFloat,
         'Other (%)': toFloat,
         'Climate': toFloat,
         'Birthrate': toFloat,
         'Deathrate': toFloat,         
         'Agriculture': toFloat,         
         'Industry': toFloat,
         'Service': toFloat
        }

df_paises = pd.read_csv('countries.csv',converters=convs)
```

### Acesso à dados no _DataFrame_

```
df_paises['Country']
```

O acesso às colunas é feito análogo à um dicionário, onde o nome da coluna desejada faz referência à chave do dicionário, o resultado é um objeto da classe _Serie_. É possível selecionar várias colunas simultaneamente, veja abaixo:

```
df_paises[['Country','Population']]
```

Nesse caso um novo objeto _DataFrame_ é retornado apenas com as colunas desejadas.
Pode-se também acessar colunas tal qual uma propriedade.

```
df_paises.Country
```

Temos duas formas de acesso à linhas do _dataframe_, a primeira forma acessando através da contagem de linhas:

```
df_paises.iloc[3]
```

Ou diretamente pelo nome no _Index_:

```
df_paises.loc[3]
```

Para facilitar o entendimento da diferença entre esse comandos vamos trabalhar um  _dataframe_ derivados dos países, mas agrupados por regiao.

```
df_paises.groupby('Region').count()
```

Acima é possível perceber que o _Index_ agora é do tipo _string_, logo para acessarmos a linha referente à Oceania por contagem, temos:

```
df_paises.groupby('Region').count().iloc[8]
```

Em poucas linhas é possível, visualmente, saber qual contagem refere-se à Ocenia, uma forma mais direta seria usar o próprio índice da linha.

```
df_paises.groupby('Region').count().loc['OCEANIA']
```

Também é possível acessar subconjuntos de linhas através de _slicing_, tanto em _iloc_, quanto em _loc_.

```
df_paises.groupby('Region').count().iloc[5:9]
```

```
df_paises.groupby('Region').count().loc[['NEAR EAST','NORTHERN AFRICA','NORTHERN AMERICA','OCEANIA']]
```

O Pandas permite a seleção de subconjuntos de dados a partir de critérios estabelecidos em forma condicionais. Por exemplo, para visualizar países cuja a população seja maior que 100M podemos fazer:

```
df_paises[df_paises.Population > 100000000]
```

Podemos usar conectivos lógicos nesse tipo de consulta. Veja os países que possuem população de mais de 100M e taxa de mortalidade menor do que 10.

```
df_paises[(df_paises.Population > 100000000) & (df_paises['Infant mortality (per 1000 births)'] < 10)]
```

### Outras funcionalidades

Podemos ordenar uma dada coluna, ou conjunto de colunas, para facilitar a visualização.

```
df_paises[['Country','Deathrate','Birthrate']].sort_values(by='Deathrate',ascending=False)
```

Novas informações podem ser extraídas dos dados, por exemplo, podemos dividir os valores das colunas _Birthrate_ por _Deathrate_ e assim obter um indicador se a população aumentou (indicador > 1),  diminuiu (indicador < 1) ou se manteve estável (indicador = 1). Tal procedimento pode ser feito por algumas vias:

```
df_paises.Birthrate.values / df_paises.Deathrate.values
```

Acima podemos ver um array do _numpy_ como resultado do cálculo que gera o indicador de crescimento populacional, apesar de estar correto, tal procedimento vem direto do uso do _numpy_ e não associa explicitamente os países aos seus respectivos indicadores. Sendo assim, podemos usar um método do Pandas chamado _apply_:

```
df_paises.apply(lambda row: row['Birthrate']/row['Deathrate'],axis=1)
```

Uma série foi retornada com os valores indexados de acordo com o _dataframe_. Sendo assim,  podemos transformar essa série numa nova coluna no _dataframe_. Veja:

```
df_paises['Pop. Growning'] = df_paises.apply(lambda row: row['Birthrate']/row['Deathrate'],axis=1)
```

E assim podemos visulizar tal coluna no _dataframe_ juntamente com todas as outras colunas.

```
df_paises[['Country','Region','Birthrate','Deathrate','Pop. Growning']]
```

Remover colunas é possível a partir do comando _drop_:

```
df_paises.drop('Pop. Growning',axis=1)
```

Entretanto, tal comando, ao invés de remover a coluna do _dataframe_ que estamos manipulando, retorna um novo _dataframe_ sem a coluna em questão. Para persistir a mudança no mesmo _dataframe_ podemos usar o parâmetro _inplace = True_ ou então atribuir o resultado para a referência do _dataframe_ em questão.

```
df_paises = df_paises.drop('Pop. Growning',axis=1)
```

### Agrupamento

O agrupamento, dado pelo comando _groupby_, é um método que implementa agrupamento por valores de colunas, da mesma forma que existe em SQL, e na estratégia __split-apply-combine__ presente na linguagem R, a qual consiste em dividir um _dataframe_ e vários através de um agrupamento, seguido da aplicação de uma computação (contagem, soma, etc.) em tais agrupamentos e por fim combina-los num único _dataframe_.

Quantos países existem por região?

```
df_paises[['Country','Region']].groupby('Region').count()
```

Lembrando que o retorno é uma espécie de _dataframe_ agrupado, nos quais os índice são os  valores presentes na coluna agrupada, no caso _Region_.

A seguir temos a estrutura de cada um dos grupos.

```
df_paises[['Country','Region']].groupby('Region').groups
```

É possível combinar métodos para melhorar a visualização dos dados.

```
df_paises[['Country','Region']].groupby('Region').count().sort_values(by='Country',ascending=False)
```

### Visualização por meio de gráficos

É possível visualizar de maneira gráfica.

```
df_paises[['Country','Region']].groupby('Region').count().sort_values(by='Country',ascending=False).plot(kind='bar',y='Country')
```

Por exemplo, desejamos saber se uma população está aumentando, ou não, para isso criaremos uma coluna que calcula _Birthrate/Deathrate_, tal qual já usamos anteriormente, entretando, nesse caso, ela será categórica, onde _Birthrate/Deathrate_ > 0 terá o valor *Grown*,  _Birthrate/Deathrate_ < 0 terá o valor *Shrink*, e  _Birthrate/Deathrate_ == 0 terá o valor *Stable*.

Logo após usaremos um gráfico de pizza para esta coluna.

```
def catGrown(row):
  return 'Grown' if row['Birthrate']/row['Deathrate'] > 1 else 'Shrink' if row['Birthrate']/row['Deathrate'] < 1 else 'Stable'

df_paises['Pop. Growning'] = df_paises.apply(catGrown,axis=1)
```

```
df_paises[['Country','Pop. Growning']].groupby('Pop. Growning').count().plot(y='Country',kind='pie')
```

Caso seja necessário visualizar o crescimento populacional por região.

```
df_paises[['Country','Region','Pop. Growning']].groupby(['Region','Pop. Growning']).count()
```

```
df_paises[['Country','Region','Pop. Growning']].groupby(['Region','Pop. Growning']).count().plot(y='Country',kind='pie')
```

Necessário transformar os índices em colunas.

```
df_region_grown = df_paises[['Country','Region','Pop. Growning']].groupby(['Region','Pop. Growning']).count()
```

```
df_region_grown.unstack(level=1).plot.bar(figsize=(10,5))
```
